```python
import matplotlib
import matplotlib.pyplot as plt
import math
```

```python
def exponent(x):
  y = []
  for entry in x:
    y.append(math.exp(entry))
  return y
```

```python
entries = 100
a = 1.0/(entries - 1.0)


x = []
for counter in range(0,entries):
  x.append(counter * a)

y = exponent(x)

plt.plot(x,y)
plt.title("CJ-7", fontweight="bold")
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.savefig("cj-7.pdf")
```
