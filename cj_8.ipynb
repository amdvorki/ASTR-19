```python
import matplotlib
import matplotlib.pyplot as plt
import math
```

```python
def sinOnList(x):
  y = []
  for entry in x:
    y.append(math.sin(entry))
  return y

def cosOnList(x):
  y = []
  for entry in x:
    y.append(math.cos(entry))
  return y
```

```python
entries = 100
a = 1.0/(entries - 1.0)


x = []
for counter in range(0,entries):
  x.append(counter * a)

sinList = sinOnList(x)
cosList = cosOnList(x)

fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.4, 0.8])
ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.8])

ax1.plot(x,sinList)
ax2.plot(x,cosList)
ax1.title.set_text("sin(x)")
ax2.title.set_text("cos(x)")
plt.savefig("cj_8.pdf")
```
